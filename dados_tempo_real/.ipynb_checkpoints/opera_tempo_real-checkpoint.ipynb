```python
import MetaTrader5 as mt5
from datetime import datetime
import pytz
import pandas as pd
from datetime import datetime
```

```python
if not mt5.initialize():
    print("initialize() failed, error code =",mt5.last_error())
    quit()
```

```python
account_info_dict = mt5.account_info()._asdict()   
# convert the dictionary into DataFrame and print
df=pd.DataFrame(list(account_info_dict.items()),columns=['property','value'])
print("account_info() as dataframe:")
print(df)
```

```python
# solicitamos 10 barras de GBPUSD D1 do dia atual
rates = mt5.copy_rates_from_pos("BBDC4", mt5.TIMEFRAME_M1, 0, 4)
  
'''# exibimos cada elemento de dados recebidos numa nova linha
print("Exibimos os dados recebidos como est√£o")
for rate in rates:
    print(rate)'''
    
# a partir dos dados recebidos criamos o DataFrame
rates_frame = pd.DataFrame(rates)
# convertemos o tempo em segundos no formato datetime
rates_frame['time']=pd.to_datetime(rates_frame['time'], unit='s')
rates_frame
```

```python
# definimos o fuso hor√°rio como UTC
timezone = pytz.timezone("Etc/UTC")
# criamos o objeto datetime no fuso hor√°rio UTC para que n√£o seja aplicado o deslocamento do fuso hor√°rio local
utc_from = datetime(2020, 8, 3, tzinfo=timezone)
# solicitamos 100 000 ticks de EURUSD a partir de 04/01/2019 no fuso hor√°rio UTC
ticks = mt5.copy_ticks_from("BBDC4", utc_from, 1500000, mt5.COPY_TICKS_ALL)
print("Ticks recebidos:",len(ticks))
 
# conclu√≠mos a conex√£o ao terminal MetaTrader 5
pd.set_option('display.max_columns', 500) # n√∫mero de colunas exibidas
pd.set_option('display.width', 1500)      # max. largura para exibi√ß√£o da tabela
 
'''# exibimos cada tick numa nova linha
print("Exibimos os ticks recebidos como s√£o")
count = 0
for tick in ticks:
    count+=1
    print(tick)
    if count >= 10:
        break'''
 
# a partir dos dados recebidos criamos o DataFrame
ticks_frame = pd.DataFrame(ticks)
# convertemos o tempo em segundos no formato datetime
ticks_frame['time']=pd.to_datetime(ticks_frame['time'], unit='s')
 
# exibimos dados
print("\nExibimos o dataframe com ticks")
ticks_frame.tail(5)
 
```

```python
import time
from IPython.display import clear_output

timezone = pytz.timezone("Etc/UTC")
utc_from = datetime(2020, 8, 3, tzinfo=timezone)

while True:
    clear_output(wait=True)
    ticks = mt5.copy_ticks_from("ITSA4", utc_from, 1500000, mt5.COPY_TICKS_ALL)
    ticks_frame = pd.DataFrame(ticks)
    ticks_frame['time']=pd.to_datetime(ticks_frame['time'], unit='s')
    print(ticks_frame.tail(5))
    time.sleep(1)
    clear()
    
```
