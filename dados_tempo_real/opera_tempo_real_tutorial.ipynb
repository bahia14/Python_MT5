```python
# import the 'pandas' module for displaying data obtained in the tabular form
pd.set_option('display.max_columns', 500) # number of columns to be displayed
pd.set_option('display.width', 1500)      # max table width to display
# import pytz module for working with time zone

# set time zone to UTC
timezone = pytz.timezone("Etc/UTC")
# create 'datetime' objects in UTC time zone to avoid the implementation of a local time zone offset
utc_from = datetime(2020, 6, 1, tzinfo=timezone)
utc_to = datetime(2020, 8, 3, tzinfo=timezone)
# request AUDUSD ticks within 11.01.2020 - 11.01.2020
ticks = mt5.copy_ticks_range("IBOV", utc_from, utc_to, mt5.COPY_TICKS_ALL)#COPY_TICKS_TRADE
print("Ticks received:",len(ticks))
 
# shut down connection to the MetaTrader 5 terminal

 
# display data on each tick on a new line
print("Display obtained ticks 'as is'")
count = 0
for tick in ticks:
    count+=1
    print(tick)
    if count >= 10:
        break
```

```python
# tentamos ativar a exibi√ß√£o do s√≠mbolo EURCAD no MarketWatch
selected=mt5.symbol_select("BIDI4",True)
if not selected:
    print("Failed to select BIDI4, error code =",mt5.last_error())
else:
    symbol_info=mt5.symbol_info("BIDI4")
    print(symbol_info)
    print("BIDI4: currency_base =",symbol_info.currency_base,"  currency_profit =",symbol_info.currency_profit,"  currency_margin =",symbol_info.currency_margin)
    print()
 
    # obtemos as propriedades do s√≠mbolo como um dicion√°rio
    print("Show symbol_info()._asdict():")
    symbol_info_dict = symbol_info._asdict()
    for prop in symbol_info_dict:
        print("  {}={}".format(prop, symbol_info_dict[prop]))
    print()
 
   # convertemos o dicion√°rio num DataFrame e imprimimos
    df=pd.DataFrame(list(symbol_info_dict.items()),columns=['property','value'])
    print("symbol_info_dict() as dataframe:")
    print(df)
```

```python
# definimos o fuso hor√°rio como UTC
timezone = pytz.timezone("Etc/UTC")
# criamos o objeto datetime no fuso hor√°rio UTC para que n√£o seja aplicado o deslocamento do fuso hor√°rio local
utc_from = datetime(2020, 8, 3, tzinfo=timezone)
# recebemos 10 barras de EURUSD H1 a partir de 01/10/2020 no fuso hor√°rio UTC
rates = mt5.copy_rates_from("VVAR3", mt5.TIMEFRAME_M5, utc_from, 10)

# exibimos cada elemento de dados recebidos numa nova linha
print("Exibimos os dados recebidos como est√£o")
for rate in rates:
    print(rate)
```

```python
# a partir dos dados recebidos criamos o DataFrame
rates_frame = pd.DataFrame(rates)
# convertemos o tempo em segundos no formato datetime
rates_frame['time']=pd.to_datetime(rates_frame['time'], unit='s')
                           
# exibimos dados
print("\nExibimos o dataframe com dados")
print(rates_frame)  
 
```

## copy_ticks_range

```python
# definimos o fuso hor√°rio como UTC
timezone = pytz.timezone("Etc/UTC")
# criamos o objeto datatime no fuso hor√°rio UTC para que n√£o seja aplicado o deslocamento do fuso hor√°rio local
utc_from = datetime(2019, 1, 1, tzinfo=timezone)
utc_to = datetime(2020, 1, 1, tzinfo=timezone)
# solicitamos ticks de AUDUSD no intervalo de 11.01.2020 - 11.01.2020
ticks = mt5.copy_ticks_range("ITSA4", utc_from, utc_to, mt5.COPY_TICKS_ALL )
print("Ticks recebidos:",len(ticks))

# a partir dos dados recebidos criamos o DataFrame
ticks_frame = pd.DataFrame(ticks)
# convertemos o tempo em segundos no formato datetime
ticks_frame['time']=pd.to_datetime(ticks_frame['time'], unit='s')
 
# exibimos dados
print("\nExibimos o dataframe com ticks")
print(ticks_frame.head(10)) 
```
